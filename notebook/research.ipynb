```python
ratings = pd.read_csv('BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={"User-ID":'User_id',
                      'Book-Rating':'Rating'}, inplace=True)
```

```python
ratings.head()
```

```python
ratings['User_id'].value_counts()
```

```python
x = ratings['User_id'].value_counts() > 200
```

```python
x
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings['User_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_rating = ratings_with_books.groupby('Title')['Rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
final_rating = ratings_with_books.merge(number_rating, on='Title')
```

```python
final_rating.head()
```

```python
# Making it so that the only books that are generated to recommend have at least 50 ratings

final_rating = final_rating[final_rating['Rating_y'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['User_id', 'Title'],inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
book_pivot = final_rating.pivot_table(columns='User_id', index='Title', values='Rating_x')
```

```python
book_pivot
```

```python
book_pivot = book_pivot.fillna(0)
```

```python
book_pivot.head()
```

```python
book_pivot.shape
```
